```python
import re

# Define the extract_numbers function
def extract_numbers(s):
    numbers = re.findall(r'\d+', s)
    return int(numbers[0]) if numbers else None

# Apply the function to the column and create a new column to store the numbers
df['imbd_title_id'] = df['imbd_title_id'].apply(extract_numbers)
```

```python
df['score'] = df['score'].apply(remove_extra_decimals)
```
